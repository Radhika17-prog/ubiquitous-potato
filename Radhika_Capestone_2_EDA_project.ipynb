## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
#Copy dataset
titles_df= titles.copy()
credits_df= credits.copy()

#Removing duplicates
titles_df.drop_duplicates(inplace=True)
credits_df.drop_duplicates(inplace=True)


````

````python

#Handling missing values
#Drop col having missing value more than 70%

titles_df.isnull().sum()/len(titles_df)*100

## seasons seems to be the col having more missing value so will drop this col
try:
  titles_df= titles_df.drop(columns = ["seasons"])
except:
  print("Col already dropped")


````

````python
titles_df.columns
````

````python
# handling missing value b/w 60 to 70% fill it with unknown and cleaning any leading/trailing whitespaces.
titles_df["age_certification"] = titles_df["age_certification"].fillna("Unknown").str.strip()
````

## Cleaning of columns

````python
#numeric conversion of runtime & release year
titles_df["runtime"] = pd.to_numeric(titles_df["runtime"], errors="coerce")
titles_df["release_year"] = pd.to_numeric(titles_df["release_year"], errors="coerce")
````

````python
 # Clean genres column – convert to list AFTER removing duplicates
titles_df["genres"] = titles_df["genres"].fillna("Unknown")
titles_df["genres_list"] = titles_df["genres"].apply(
    lambda x: x.split(", ") if x != "Unknown" else []
)
````

````python
#Clean production countries column
titles_df["production_countries"] = titles_df["production_countries"].fillna("Unknown")
titles_df["countries_list"] = titles_df["production_countries"].apply(
    lambda x: x.split(", ") if x != "Unknown" else []
)
````

````python
titles_df
````

#Credit table cleaning

````python
#Changing datatype
credits_df["role"] = credits_df["role"].astype(str)
credits_df["name"] = credits_df["name"].astype(str)
````

````python
titles_df.head()
````

### Clean Imdb-col

````python
# Clean Imdb-col
titles_df["imdb_id"] = titles_df["imdb_id"].fillna('Unknown').str.strip()
````

````python
titles_df['imdb_id'].isnull().sum()
````

#Clean imdb score

````python
titles_df['imdb_score'].dtype
````

#Outlier removal based on skewness

````python

num_cols = ["runtime", "imdb_score", "imdb_votes", "tmdb_popularity","tmdb_score"]

for col in num_cols:
    if col in titles_df.columns:
        Q1 = titles_df[col].quantile(0.25)
        Q3 = titles_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        titles_df = titles_df[(titles_df[col] >= lower) & (titles_df[col] <= upper)]

titles_df.head()
````
